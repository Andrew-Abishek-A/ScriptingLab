```python
import numpy as np
import pandas as pd


studentperf_df = pd.read_csv("StudentsPerformance.csv")

studentperf_df['test preparation course'] = studentperf_df['test preparation course'].map({
    'none' : 'Not completed',
    'completed' : 'Completed'
})

studentperf_df['race/ethnicity'] = studentperf_df['race/ethnicity'].map({
    'group A' : 'Asian Students',
    'group B' : 'African Students',
    'group C' : 'Afro-Asian Students',
    'group D' : 'American Students',
    'group E' : 'European Students',
})

studentperf_df['parental level of education'] = studentperf_df['parental level of education'].fillna('not applicable')

studentperf_1_df = studentperf_df.drop(['parental level of education', 'test preparation course', 'lunch'],\
                                       axis = 1, inplace = False)

studentperf_2_df = studentperf_df.drop(['parental level of education', 'test preparation course', 'lunch', 'gender',\
                                        'race/ethnicity'], axis = 1, inplace = False)


print(studentperf_1_df.head(5))
#print(studentperf_2_df.head(5))
# print(studentperf_df)

import seaborn as sb
import matplotlib.pyplot as plt


ax = sb.countplot(x = 'test preparation course', data = studentperf_df, hue = 'gender', palette = 'Set1')\
 .set(xlabel = 'Groups', ylabel = 'Total', Title = 'No of students per group')

plt.show()


ax = sb.countplot(x = 'race/ethnicity', data = studentperf_df, hue = 'gender', palette = 'Set2')\
.set(xlabel = 'Groups', ylabel = 'Total', Title = 'No of students per group')
plt.xticks(rotation=45, ha="right")
plt.show()


intervals = [0, 40, 50, 60, 70, 80, 90, 100]
grades = ['F', 'E', 'D', 'C', 'B', 'A', 'S']

studentperf_df['grades'] = pd.cut(studentperf_2_df.mean(axis = 1), intervals, labels = grades)


# ax = sb.countplot(x = '', data = studentperf_df, hue = 'grades', palette = 'Set1')\
#  .set(xlabel = 'x axis', ylabel = 'y axis', Title = 'Title')

# plt.show()


ax = sb.countplot(x = 'gender', data = studentperf_df, hue = 'grades', palette = 'Set1')\
 .set(xlabel = 'x axis', ylabel = 'y axis', Title = 'Title')

plt.show()


# ax = sb.countplot(x = 'race/ethnicity', data = studentperf_df, hue = 'grades', palette = 'Set2')\
#  .set(xlabel = 'x axis', ylabel = 'y axis', Title = 'Title')

# plt.show()

# ax = sb.countplot(x = 'parental level of education', data = studentperf_df, hue = 'grades', palette = 'Set3')\
#  .set(xlabel = 'x axis', ylabel = 'y axis', Title = 'Title')

# plt.show()
```
